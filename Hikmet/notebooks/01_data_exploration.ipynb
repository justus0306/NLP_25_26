```python
#download dataset from huggingface
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("cardiffnlp/tweet_topic_multi")
print(ds)
```

```python
#imports pandas-library and pulls dataset(ds) into dataframe
import pandas as pd
df = pd.DataFrame(ds["train_all"]) #convert to pandas dataframe
df.head() #show first 5 rows of dataframe
```

```python
df["label_name"].value_counts()
```

```python
import matplotlib.pyplot as plt

# Top 20 Themen nach Häufigkeit
top_labels = df["label_name"].value_counts().head(20)

top_labels.plot(kind="bar", figsize=(12, 6))
plt.title("Top 20 Themen in train_all")
plt.xlabel("Thema")
plt.ylabel("Anzahl Tweets")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.savefig("../results/themenverteilung.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Top 20 Themen nach Häufigkeit
top_labels = df["label_name"].value_counts().head(20)

# Horizontales Balkendiagramm
top_labels.plot(kind="barh", figsize=(10, 8), color="skyblue")
plt.title("Top 20 Themen in train_all")
plt.xlabel("Anzahl Tweets")
plt.ylabel("Thema")
plt.gca().invert_yaxis()  # Wichtig: damit das häufigste Thema oben steht
plt.tight_layout()

plt.savefig("../results/themenverteilung.png", dpi=300, bbox_inches='tight')
plt.show()

```
